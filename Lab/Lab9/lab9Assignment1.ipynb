```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y1 = x**2
y2 = x**3

plt.subplot(211)
plt.plot(x, y1, 'r--')
plt.title('Lab Assignment1: x^2 and x^3 using subplot() method')
plt.xlabel('x')
plt.ylabel('x^2')
plt.subplot(212)
plt.plot(x, y2, 'g-.')
plt.xlabel('x')
plt.ylabel('x^3')
```

```python
fig, axs = plt.subplots(2)
fig.suptitle('Lab Assignment1: x^2 and x^3 using subplots() method')
axs[0].plot(x, y1, 'b-')
axs[0].set_title('x^2')
axs[0].set_xlabel('x')
axs[0].set_ylabel('x^2')
axs[1].plot(x, y2, 'm:')
axs[1].set_title('x^3')
axs[1].set_xlabel('x')
axs[1].set_ylabel('x^3')
```
