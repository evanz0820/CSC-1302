```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fig = plt.figure()
axe = fig.add_axes([.5, .5, 1, 1])
x = np.arange(0,np.pi*2, .1)
y1 = np.sin(x)
y2 = np.cos(x)

axe.plot(x, y1, 'r-.*')
axe.set_title('Axe object')
axe.set_xlabel('X Label')
axe.set_ylabel('Y Label')
axe.legend(labels = 'sinx', loc = 1)
```

```python
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]

fig1 = plt.figure()

axe1 = fig1.add_axes([0, 0, 1, 1])
axe1.plot(x1, y, 'r-.*')
axe1.plot(x2, y, 'g--o')
axe1.set_title('Advertisement effect on sales')
axe1.set_xlabel('Medium')
axe1.set_ylabel('Sales')
axe1.legend(labels = ('TV', 'SmartPhone'), loc = 0)
```

```python
x1 = np.linspace(-5, 5, 50)
y1 = 2 * x1
y2 = x1 ** 2 - 10

fig4 = plt.figure()

axe4 = fig4.add_subplot(2,1,1)
axe4.plot(x1, y1)

axe5 = fig4.add_subplot(2,1,2)
axe5.plot(x1, y2)

plt.suptitle('Multiple Plots')
plt.show()
```

```python
fig5, axe6 = plt.subplots(1,2)

axe6[0].plot(x1, y1)
axe6[1].plot(x1, y2)

plt.suptitle('Subplots')
```
