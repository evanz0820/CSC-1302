```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,8])
y = np.array([3,10])

plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Example Diagram')

plt.plot(x,y, 'o--r', markersize = 3, linewidth = 2, label = 'x_line')
plt.legend()
plt.show()
```

```python
year = [1972,1982,1992,2002,2012]
e_india = [100.6, 158.61, 305.54, 394.96, 724.79]
e_bangladesh = [10.5, 25.21, 58.65, 119.27, 274.87]
```

```python
plt.xlabel('Year')
plt.ylabel('Electricity Consumption')
plt.title('Electricity Consumption vs Year')

plt.plot(year, e_india, '--b', markersize = 50, linewidth = 6, label = 'e_india')
plt.plot(year, e_bangladesh, ':y', markersize = 50, linewidth = 3, label = 'e_bangledash')
plt.legend()
plt.show()
```

```python
x = np.linspace(-5,5,50)
y1 = 2 * x
y2 = x ** 2
plt.plot(x,y1)
plt.plot(x, y2)
plt.xlim((1980, 2000))
plt.ylim((0, 500))
plt.xticks([1985, 1990, 1995], ['x1', 'x2', 'x3'])
plt.yticks([0,100,200,300,400], ['y1', 'y2', 'y3', 'y4', 'y5'])
plt.legend()
plt.show()
```
